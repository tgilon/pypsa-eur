```python
import pypsa
import pandas as pd
```

# Nuclear kept from add_electricity

```python
n_elec = pypsa.Network("results/test-elec/networks/base_s_5_elec_.nc")
n_sector_myopic_2030 = pypsa.Network("results/test-sector-myopic/networks/base_s_5___2030.nc")
n_sector_myopic_2040 = pypsa.Network("results/test-sector-myopic/networks/base_s_5___2040.nc")
n_sector_myopic_2050 = pypsa.Network("results/test-sector-myopic/networks/base_s_5___2050.nc")
n_sector_overnight = pypsa.Network("results/test-sector-overnight/networks/base_s_5___2030.nc")
n_sector_perfect = pypsa.Network("results/test-sector-perfect/networks/base_s_5___brownfield_all_years.nc")
```

```python
pd.concat([
    n_elec.statistics.optimal_capacity(bus_carrier="AC").filter(like="nuclear").to_frame(name="elec"),
    n_sector_myopic_2030.statistics.optimal_capacity(bus_carrier="AC").filter(like="nuclear").to_frame(name="myopic"),
    n_sector_myopic_2040.statistics.optimal_capacity(bus_carrier="AC").filter(like="nuclear").to_frame(name="myopic"),
    n_sector_myopic_2050.statistics.optimal_capacity(bus_carrier="AC").filter(like="nuclear").to_frame(name="myopic"),
    n_sector_overnight.statistics.optimal_capacity(bus_carrier="AC").filter(like="nuclear").to_frame(name="overnight"),
    n_sector_perfect.statistics.optimal_capacity(bus_carrier="AC").filter(like="nuclear",axis=0),
]).fillna('-')
```

```python
pd.concat([
    n_elec.statistics.installed_capacity(bus_carrier="AC").filter(like="nuclear").to_frame(name="elec"),
    n_sector_myopic_2030.statistics.installed_capacity(bus_carrier="AC").filter(like="nuclear").to_frame(name="myopic"),
    n_sector_myopic_2040.statistics.installed_capacity(bus_carrier="AC").filter(like="nuclear").to_frame(name="myopic"),
    n_sector_myopic_2050.statistics.installed_capacity(bus_carrier="AC").filter(like="nuclear").to_frame(name="myopic"),
    n_sector_overnight.statistics.installed_capacity(bus_carrier="AC").filter(like="nuclear").to_frame(name="overnight"),
    n_sector_perfect.statistics.installed_capacity(bus_carrier="AC").filter(like="nuclear",axis=0),
]).fillna('-')
```

```python
pd.concat(
    [
        pd.concat([n_elec.generators.filter(like="nuclear",axis=0)[["p_max_pu", "p_min_pu"]]], axis=1, keys=["elec"]),
        pd.concat([n_elec.links.filter(like="nuclear",axis=0)[["p_max_pu", "p_min_pu"]]], axis=1, keys=["elec"]),
        pd.concat([n_sector_myopic_2030.generators.filter(like="nuclear",axis=0)[["p_max_pu", "p_min_pu"]]], axis=1, keys=["myopic"]),
        pd.concat([n_sector_myopic_2030.links.filter(like="nuclear",axis=0)[["p_max_pu", "p_min_pu"]]], axis=1, keys=["myopic"]),
        pd.concat([n_sector_myopic_2040.generators.filter(like="nuclear",axis=0)[["p_max_pu", "p_min_pu"]]], axis=1, keys=["myopic"]),
        pd.concat([n_sector_myopic_2040.links.filter(like="nuclear",axis=0)[["p_max_pu", "p_min_pu"]]], axis=1, keys=["myopic"]),
        pd.concat([n_sector_myopic_2050.generators.filter(like="nuclear",axis=0)[["p_max_pu", "p_min_pu"]]], axis=1, keys=["myopic"]),
        pd.concat([n_sector_myopic_2050.links.filter(like="nuclear",axis=0)[["p_max_pu", "p_min_pu"]]], axis=1, keys=["myopic"]),
        pd.concat([n_sector_overnight.generators.filter(like="nuclear",axis=0)[["p_max_pu", "p_min_pu"]]], axis=1, keys=["overnight"]),
        pd.concat([n_sector_overnight.links.filter(like="nuclear",axis=0)[["p_max_pu", "p_min_pu"]]], axis=1, keys=["overnight"]),
        pd.concat([n_sector_perfect.generators.filter(like="nuclear",axis=0)[["p_max_pu", "p_min_pu"]]], axis=1, keys=["perfect"]),
        pd.concat([n_sector_perfect.links.filter(like="nuclear",axis=0)[["p_max_pu", "p_min_pu"]]], axis=1, keys=["perfect"]),
    ]
).fillna("-")
```
